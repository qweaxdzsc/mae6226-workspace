# Source & Sink

# Overview

The theoretical basis for classical Aerodynamics is potential-flow theory

assumptions:
    #the flow is steady;
    #the velocity remains smaller than the speed of sound (incompressible flow);
    #the fluid has no internal friction, i.e., is inviscid; and
    #it has no vorticity (fluid particles are not rotating).
#and this theory has a special mathematical quality: linear

the governing equation for the potential flow: $$\nabla^2\phi=0$$

# Start

```python
import numpy as np
import math
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
#generates a mesh grid
N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
x = np.linspace(x_start, x_end, N)
y = np.linspace(y_start, y_end, N)
X, Y = np.meshgrid(x, y)
print(X, Y)
```

```python
%matplotlib inline
```

```python
# plot the grid of points
size = 10
plt.figure(figsize=(size, size*(y_end-y_start)/(x_end-x_start)))
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('grid of points', fontsize=30)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.scatter(X, Y, s=10, color='#CD2305', marker='o')
```

# Source flow

```python
strength_source = 10.0
x_source, y_source = -1.0, 0.0
u_source = strength_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
```

# Sink flow

```python
strength_sink = -5.0                     
x_sink, y_sink = 1.0, 0.0                
u_sink = strength_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_sink, y_sink, color='#CD2305', s=80, marker='o', linewidth=0);
```

# Source-sink pair

```python
u_pair = u_source + u_sink
v_pair = v_source + v_sink
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_pair, v_pair, density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], 
            color='#CD2305', s=80, marker='o', linewidth=0);
```
