task 1 creat function

```python
def KT_transform(zeta, b, n):
    T1 = (1 + b**2/zeta)**n
    T2 = (1 - b**2/zeta)**n
    z = n * (T1 + T2)/(T1 - T2)
    return z
```

to check

```python
zeta = 2.2 + 2.2*1j
b = 1
theta = 5/180*numpy.pi
n = 2- theta/numpy.pi
Z = KT_transform(zeta, b, n)
print(Z)
```
